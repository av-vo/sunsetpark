## PWMsg
*Work with PulseWaves data in PWMsg Protocol Buffers format*

```python
from als.pulsewaves import PulseWaves
```

#### Initiate a PulseWaves object

```python
# input pwmsg file
pwmsg_fname = '../data/sample/F_150326_115601-0.pwmsg'

# initiate a PulseWaves object
pw = PulseWaves(pwmsg_fname)

# print out the scale matrix
pw.scale
```

#### Access a pulse

```python
pulse = pw.get_pulse(pulse_idx = 0)
pulse
```

#### Access a sampling (aka channel)

```python
sampling = pw.get_sampling(pulse_idx = 0, sampling_idx = 1)
sampling
```

#### Access a wave segment

```python
segment = pw.get_segment(pulse_idx = 0, sampling_idx = 0, segment_idx = 0)
segment
```

#### Access waveform samples of a segment

Get the stored samples without georeferencing

```python
samples = pw.get_samples(pulse_idx = 0, sampling_idx = 0, segment_idx = 0, georef = False)
samples
```

Georeference the samples

```python
samples = pw.get_samples(pulse_idx = 0, sampling_idx = 0, segment_idx = 0, georef = True)
samples
```

#### Plot the waveforms of a pulse

Plot waveform amplitude against index (no referencing required)

```python
fig, axes = pw.plot_waves(pulse = 0, georef = False)
```

Plot waveform amplitude against elevation (geoferencing required)

```python
fig, axes = pw.plot_waves(pulse = 0, georef = True)
```
