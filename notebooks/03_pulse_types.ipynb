```python
import pandas as pd

from als.pulsewaves import PulseWaves
```

### Get data

```python
# input pwmsg file
pwmsg_fname = '../data/sample/10552_NYU_M2_Pulse_Waves_MTA_Resolved-Scanner1-190511_161348_1-originalpoints-0.pwmsg'

# initiate a PulseWaves object
pw = PulseWaves(pwmsg_fname)
```

### Different pulse types

```python
pulse_list = pw.pwmsg.pulse

desc_idx = pd.Series([pls.pulseDescriptorIndex for pls in pulse_list])
print('pulse descriptor indices: {0}'.format(desc_idx.unique()))

tt = 0
for n in desc_idx.unique():
    idx=desc_idx.index[desc_idx==n]
    print('number of pulses with pulse descriptor #{0}: {1}'.format(n, len(idx)))
    tt += len(idx)
```

#### Type 2
1 x RP, 1 x LP, 0 x HP

```python
idx = 2
idx=desc_idx.index[desc_idx==idx]
pls = pulse_list[idx[7]]
#print(pls) # uncomment this if you want to read the pulse as JSON
fig, axes = pw.plot_waves(pls)
```
